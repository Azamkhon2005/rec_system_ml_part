```python
import numpy as np
import pandas as pd
```

```python
# метод для извлечение данных из датасетов
def import_data(num):
    path = f"data\{num}.xlsx"
    data = pd.read_excel(path, sheet_name = 'Результаты поиска')
    data.columns = data.loc[0]
    data = data.drop([0])
    print(data.shape)
    return data
```

```python
# соединение данных в один датафрейм
arr = []
for i in range(1, 24):
    prep = import_data(i)
    arr.append(prep)
data = pd.concat(arr)
```

```python
data
```

```python
# удаление записей, в которых не присутствует данные поставщика
data = data[(data['Поставщик'] != 'Неизвестно') & (data['Поставщик'] != 'Несколько поставщиков')].drop_duplicates()
```

```python
# все столбцы датафрейма
data.columns
```

```python
# удаление записей, в которых не присутствует данные поставщика
data = data[data['Поставщик'].isnull()==False]
```

```python
# сохранение единого датафрейма в excel таблицу
data.to_excel('cleared_data.xlsx')
```
